# 4. Produtos Mais e Menos Vendidos

```python
# Loja 1
# - Criar dicionario para os produtos
produtos_loja1 = {}

for produto in loja['Produto']:
    if produto in produtos_loja1:
        produtos_loja1[produto] += 1
    else:
        produtos_loja1[produto] = 1

# - Função para identificar o produto mais vendido
def identificar_produtos_mais_vendidos(dicionario_produtos):
    maior_valor = max(dicionario_produtos.values())
    produtos_maior_empate = []

    for produto, qtd in dicionario_produtos.items():
        if qtd == maior_valor:
            produtos_maior_empate.append(produto)

    return produtos_maior_empate, maior_valor

# - Função para identificar o produto menos vendido
def identificar_produtos_menos_vendidos(dicionario_produtos):
    menor_valor = min(dicionario_produtos.values())
    produtos_menor_empate = []

    for produto, qtd in dicionario_produtos.items():
        if qtd == menor_valor:
            produtos_menor_empate.append(produto)

    return produtos_menor_empate, menor_valor

produtos_mais_vendidos_loja1 = identificar_produtos_mais_vendidos(produtos_loja1)
produtos_menos_vendidos_loja1 = identificar_produtos_menos_vendidos(produtos_loja1)
```

```python
# loja 2
produtos_loja2 = {}

for produto in loja2['Produto']:
    if produto in produtos_loja2:
        produtos_loja2[produto] += 1
    else:
        produtos_loja2[produto] = 1

produtos_mais_vendidos_loja2 = identificar_produtos_mais_vendidos(produtos_loja2)
produtos_menos_vendidos_loja2 = identificar_produtos_menos_vendidos(produtos_loja2)
```

```python
# loja 3
produtos_loja3 = {}

for produto in loja3['Produto']:
    if produto in produtos_loja3:
        produtos_loja3[produto] += 1
    else:
        produtos_loja3[produto] = 1

produtos_mais_vendidos_loja3 = identificar_produtos_mais_vendidos(produtos_loja3)
produtos_menos_vendidos_loja3 = identificar_produtos_menos_vendidos(produtos_loja3)
```

```python
# loja 4
produtos_loja4 = {}

for produto in loja4['Produto']:
    if produto in produtos_loja4:
        produtos_loja4[produto] += 1
    else:
        produtos_loja4[produto] = 1

produtos_mais_vendidos_loja4 = identificar_produtos_mais_vendidos(produtos_loja4)
produtos_menos_vendidos_loja4 = identificar_produtos_menos_vendidos(produtos_loja4)
```

```python
# Mostrar produtos mais vendidos de cada loja
print(f'O(s) produto(s) mais vendido(s) na loja 1: {produtos_mais_vendidos_loja1}')
print(f'O(s) produto(s) mais vendido(s) na loja 2: {produtos_mais_vendidos_loja2}')
print(f'O(s) produto(s) mais vendido(s) na loja 3: {produtos_mais_vendidos_loja3}')
print(f'O(s) produto(s) mais vendido(s) na loja 4: {produtos_mais_vendidos_loja4}')

# Mostrar produtos menos vendidos de cada loja
print(f'O(s) produto(s) menos vendido(s) na loja 1: {produtos_menos_vendidos_loja1}')
print(f'O(s) produto(s) menos vendido(s) na loja 2: {produtos_menos_vendidos_loja2}')
print(f'O(s) produto(s) menos vendido(s) na loja 3: {produtos_menos_vendidos_loja3}')
print(f'O(s) produto(s) menos vendido(s) na loja 4: {produtos_menos_vendidos_loja4}')
```
